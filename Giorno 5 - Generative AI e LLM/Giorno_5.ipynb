## Esempio 1: Generazione di testo

### GPT2

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
MODEL_NAME = "gpt2"
```

```python
tokenizer = GPT2Tokenizer.from_pretrained(MODEL_NAME)
```

```python
PROMPT = "How bad is a cat ?"

# Tokenize input
input_ids = tokenizer.encode(PROMPT, return_tensors="pt", add_special_tokens=True)
input_ids
```

```python
model = GPT2LMHeadModel.from_pretrained(MODEL_NAME)
```

```python
output = model.generate(input_ids, max_length=100, num_return_sequences=1, no_repeat_ngram_size=2, top_k=50, top_p=0.95, temperature=0.7)
output
```

```python
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

## Esempio 2: Generazione di codice Python
 - [Repository su Hugging Face](https://huggingface.co/codellama/CodeLlama-7b-hf)

```python
!pip install accelerate
```

```python
from transformers import AutoTokenizer
import transformers
import torch
```

```python
MODEL_NAME = "codellama/CodeLlama-7b-hf"
```

```python
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model=MODEL_NAME,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
prompt = "def hello_world():"
```

```python
def generate_code(prompt):
    output = pipeline(
        prompt,
        do_sample=True,
        temperature=0.1,
        top_p=0.9,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=128,
    )
    print(output[0]["generated_text"])


generate_code(prompt)
```

## Esempio 3: Generazione di traduzioni
 - [Repository su HuggingFace](https://huggingface.co/Helsinki-NLP/opus-mt-it-en)
 - [Documentazione MarianMTModel](https://huggingface.co/docs/transformers/model_doc/marian)

```python
from transformers import AutoTokenizer, MarianMTModel
```

```python
SOURCE = "it"
TARGET = "en"

MODEL_NAME = f"Helsinki-NLP/opus-mt-{SOURCE}-{TARGET}"
```

```python
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = MarianMTModel.from_pretrained(MODEL_NAME)
```

```python
prompt = "I gatti sono cattive persone"
```

```python
input_ids = tokenizer.encode(PROMPT, return_tensors="pt")
output = model.generate(input_ids)
translation = tokenizer.decode(output[0], skip_special_tokens=True)
print(translation)
```

```python
batch = tokenizer([prompt], return_tensors="pt")
generated_ids = model.generate(**batch)
tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
```
